## Question 12: Get team long name with total of goal, order the total number by descending

- Total of goal of a team is added up from both away and home games
- Use UNION statement 

```python
#write your query
result = pd.read_sql('''select X.team_long_name, sum(X.sum_home) as total_goal
                          from (select T1.team_long_name, sum(M1.home_team_goal) as sum_home
                                from Match as M1  
                                inner join team as T1 on M1.home_team_api_id = T1.team_api_id
                                group by T1.team_api_id
                                union
                                select T2.team_long_name, sum(M2.away_team_goal) as sum_away
                                from Match as M2 
                                inner join team as T2 on M2.away_team_api_id = T2.team_api_id
                                group by T2.team_api_id) as X
                      group by X.team_long_name
                      order by total_goal desc''', conn)
result
```

## Question 13: Get team long name with total of matches, order the total number by descending

```python
#write your query
result = pd.read_sql('''select team_long_name, sum(count_team_match) as total_match
                        from (select T1.team_long_name as team_long_name, count(*) as count_team_match
                              from Match as M1  
                              inner join team as T1 on M1.home_team_api_id = T1.team_api_id
                              group by T1.team_long_name
                              union all
                              select T2.team_long_name as team_long_name, count(*) as count_team_match
                              from Match as M2
                              inner join team as T2 on M2.away_team_api_id = T2.team_api_id
                              group by T2.team_long_name) as X
                        group by team_long_name
                        order by total_match desc''', conn)
result
```

## Question 14: Get numbers of win, lost and draw matches of each team

- If a team has home goal > away goal , team is win in this match
- If a team has home goal < away goal , team is lost in this match
- If a team has home goal = away goal , team is draw in this match
- Use Case When statement


```python
#write your query
result = pd.read_sql('''select team_long_name, sum(win) as win, sum(lose) as lose, sum(draw) as draw
                        from (select team_long_name,
                              case
                                  when home_team_goal > away_team_goal then '1'
                                  else '0'
                              end as win,
                              case
                                  when home_team_goal < away_team_goal then '1'
                                  else '0'
                              end as lose,
                              case
                                  when home_team_goal = away_team_goal then '1'
                                  else '0'
                              end as draw
                              from Match as M1  
                              inner join team as T1 on M1.home_team_api_id = T1.team_api_id
                              union all
                              select team_long_name,
                              case
                                when home_team_goal < away_team_goal then '1'
                                else '0'
                              end as win,
                              case
                                when home_team_goal > away_team_goal then '1'
                                else '0'
                              end as lose,
                              case
                                when home_team_goal = away_team_goal then '1'
                                else '0'
                              end as draw      
                              from Match as M2 
                              inner join team as T2 on M2.away_team_api_id = T2.team_api_id
                              ) as X
                        group by team_long_name
                        order by win desc''', conn)
result
```

## Question 15: Get top 10 team with highest numbers of win matches

## Question 17: Get percentage of score in each league
